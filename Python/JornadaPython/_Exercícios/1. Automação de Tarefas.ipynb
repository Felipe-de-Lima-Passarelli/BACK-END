```python
#Passo 1: importar as bibliotecas necessárias para registar o banco de dados automaticamente
import pyautogui
import pandas
from time import sleep
```

```python
#Passo 2: Importar o banco de dados
tabela = pandas.read_csv("Dados/Produtos.csv")
display(tabela) #Possui 293 produtos
```

```python
#Passo 3: Automatizar o registro dos produtos
pyautogui.PAUSE = 0.5

#Abrir navegador
pyautogui.press("win")
pyautogui.write("op")
pyautogui.press("enter")
sleep(0.5)

#Entrar no site
link = "https://dlp.hashtagtreinamentos.com/python/intensivao/login"
pyautogui.write(link)
pyautogui.press("enter")
sleep(0.5)

#Logar no site
pyautogui.click(x=799, y=393)
pyautogui.write("Email_ficticio@gmail.com")
pyautogui.press("tab")
pyautogui.write("Senha_ficticia")
pyautogui.press("tab")
pyautogui.press("enter")
sleep(0.5)

#Registrar todos produtos automaticamente
for linha in tabela.index:
    pyautogui.click(x=788, y=277)
    codigo = tabela.loc[linha, "codigo"]
    pyautogui.write(str(codigo))
    pyautogui.press("tab")

    marca = tabela.loc[linha, "marca"]
    pyautogui.write(str(marca))
    pyautogui.press("tab")

    tipo = tabela.loc[linha, "tipo"]
    pyautogui.write(str(tipo))
    pyautogui.press("tab")

    categoria = tabela.loc[linha, "categoria"]
    pyautogui.write(str(categoria))
    pyautogui.press("tab")

    preco = tabela.loc[linha, "preco_unitario"]
    pyautogui.write(str(preco))
    pyautogui.press("tab")

    custo = tabela.loc[linha, "custo"]
    pyautogui.write(str(custo))
    pyautogui.press("tab")

    obs = tabela.loc[linha, "obs"]
    if not pandas.isna(obs):
        pyautogui.write(str(obs))
    pyautogui.press("tab")
    pyautogui.press("enter")
    pyautogui.scroll(1000)
```
